This notebook shows the distribution of angle and throttle

The fist example shows the angle

```python
from numpy import arange
from pylab import plot,show
from pathlib import Path
import os
import json
import glob

sample_size = 1000
x = arange(0,sample_size)
y = []

tubPath = Path('/Users/dan/Documents/Projects/DonkeyCar/tub/test-set-01')
json_files = glob.glob1(tubPath,"record_*.json")
sample_files = json_files[:sample_size]

for filename in sample_files:
    data = json.load(open(os.path.join(tubPath, filename)))
    y.append(data["user/angle"])
    
plot(x,y, '.')
show()
```

Note that because the track is curcular, most of the turns are to the right.  I should also drive counter clockwise.

Now let's look at the throttle.  The car is usally going forward, so the data is mostly forward.  There are a few backward numbers.


```python
from numpy import arange
from pylab import plot,show
from pathlib import Path
import os
import json
import glob
from pprint import pprint

sample_size = 1000
x = arange(0,sample_size)
y = []

tubPath = Path('/Users/dan/Documents/Projects/DonkeyCar/tub/test-set-01')
json_files = glob.glob1(tubPath,"record_*.json")
sample_files = json_files[:sample_size]

for filename in sample_files:
    data = json.load(open(os.path.join(tubPath, filename)))
    y.append(data["user/throttle"])
    
plot(x,y, '.')
show()
```
